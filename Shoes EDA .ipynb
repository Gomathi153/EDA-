## BoxPlot

```python
plt.boxplot(df.reviews)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.catplot(x="discount", y="listing_price", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for listing_price vs reviews")
plt.show()
```

```python
sns.catplot(x="rating", y="reviews", data=df, kind="box", aspect=1.3)
plt.title("Boxplot for rating vs reviews")
plt.show()
```

```python
sns.scatterplot(x="listing_price", y="sale_price", hue="reviews", data=df, palette="Dark2", s=80)
plt.title("Relationship between listing_price, sale_price and reviews")
plt.show()
```

```python
sh = sns.heatmap(df[['listing_price','rating']].corr(),annot=True)
```

```python
corrmat = df.corr()
```

```python
sns.heatmap(corrmat,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size': 10},yticklabels=df.columns,xticklabels=df.columns,cmap='Spectral_r')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,6)
sns.heatmap(df.isnull(),yticklabels=False, cbar = False, cmap='viridis')
plt.title('Missing null values')
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.kdeplot(df.rating, color='Violet', shade = True)
plt.xlabel("listing_price")
plt.ylabel("sale_price")
plt.title("Listing and Sale prices", size = 20)
```

```python
plt.figure(figsize=(8,8))
plt.title('Sale_price and Reviews')
sns.scatterplot(x='sale_price',y='reviews',data=df,hue='listing_price')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),cbar = True, cmap = 'coolwarm')
```

```python
plt.figure(figsize=(8,6))
data = df.groupby('listing_price')['sale_price'].max().sort_values(ascending = True)
data = data.head()
labels = data.keys()
plt.pie(data, labels = labels, autopct = '%.0f%%')
plt.title('Top 10 sale in listing price', fontsize=15)
```

```python
sns.jointplot(x='listing_price',y='sale_price',data=df,kind='reg')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
sns.heatmap(corr,cmap=['violet','yellow','pink'],vmin = -.5, vmax = 0.6, annot = True)
plt.show()
```

```python
sns.displot(x='sale_price', kde = True, bins = 5, hue = df['rating'], palette = 'Set3',data = df)
plt.show()
```
